### Colab Workspace Setup

```python
import os
import subprocess

BASE_DIR = "/content/MAIE5531-S2026---GEN-AI-and-LLM-main"
ASSIGNMENT2_DIR = os.path.join(BASE_DIR, "assignment2")
LLAMA_DIR = os.path.join(ASSIGNMENT2_DIR, "llama_training")


def run_bash_script(script_name, working_dir):
    working_dir = os.path.abspath(working_dir)
    script_path = os.path.join(working_dir, script_name)

    print(f"Running: {script_name}")
    print(f"Working directory: {working_dir}")

    if not os.path.exists(working_dir):
        raise FileNotFoundError(f"Directory does not exist: {working_dir}")
    if not os.path.exists(script_path):
        raise FileNotFoundError(f"Script does not exist: {script_path}")

    subprocess.run(["bash", script_name], cwd=working_dir, check=True)


print("Colab paths configured.")
print(f"Base dir: {BASE_DIR}")
print(f"Assignment 2 dir: {ASSIGNMENT2_DIR}")
print(f"LLaMA training dir: {LLAMA_DIR}")
```

### Clone Codebase

```python
import os
import subprocess
import sys

# 你的仓库地址
REPO_URL = "https://github.com/Sylvia2001vck/MAIE5531-S2026---GEN-AI-and-LLM--hw2.git"
BASE_DIR = "/content/MAIE5531-HW2"

# 1. 克隆仓库（如果已存在就跳过）
if not os.path.exists(BASE_DIR):
    subprocess.run(["git", "clone", REPO_URL, BASE_DIR], check=True)
    print("✅ 仓库克隆完成")
else:
    print("✅ 仓库已存在")

# 2. 【关键修复】补全嵌套的目录层级！
MAIN_REPO_DIR = os.path.join(BASE_DIR, "MAIE5531-S2026---GEN-AI-and-LLM-main")
ASSIGNMENT2_DIR = os.path.join(MAIN_REPO_DIR, "assignment2")
LLAMA_DIR = os.path.join(ASSIGNMENT2_DIR, "llama_training")

# 3. 打印目录，确认脚本存在
print("\n📂 目录结构检查：")
print(f"主目录: {MAIN_REPO_DIR}")
print(f"作业目录: {ASSIGNMENT2_DIR}")
print(f"训练目录: {LLAMA_DIR}")
print(f"\n✅ 作业目录下的文件: {os.listdir(ASSIGNMENT2_DIR)}")  # 会显示 download_data.sh

# 4. 运行脚本函数
def run_bash_script(script_name, working_dir):
    print(f"\n=====================================")
    print(f"Running: {script_name}")
    print(f"Directory: {working_dir}")

    script_path = os.path.join(working_dir, script_name)
    if not os.path.exists(script_path):
        print(f"❌ 脚本不存在: {script_path}")
        print(f"目录文件列表: {os.listdir(working_dir)}")
        return

    # Colab 专用：实时输出，不缓存，不卡死
    import sys
    process = subprocess.Popen(
        ["bash", script_name],
        cwd=working_dir,
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        bufsize=1,
        universal_newlines=True
    )

    # 逐行实时打印
    for line in iter(process.stdout.readline, ''):
        sys.stdout.write(line)
        sys.stdout.flush()

    process.wait()
    print(f"\n✅ 脚本执行完成，退出码: {process.returncode}")
```
